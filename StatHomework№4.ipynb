**3)Непрерывная случайная величина X распределена нормально и задана плотностью распределения** $f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32)$

**Найдите:**

а) M(X)

б) D(X)

в) std(X) (среднее квадратичное отклонение)

Нормальное распределение задается функцией плотности вида:
$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right) $

Из условий задачи:

$ f(x) = \frac{1}{4 \sqrt{2\pi}} \exp\left(-\frac{(x+2)^2}{32}\right) $

Сравнивая с общей формулой нормального распределения, мы видим, что:

$\sigma = 4$

$\mu = -2$

аким образом, математическое ожидание $M(X) равно:

$M(X) = \mu = -2$

б) D(X) (Дисперсия)
Дисперсия $D(X)$ для нормального распределения равна квадрату стандартного отклонения $σ$:

$D(X) = σ^2 = 4^2 = 16$

в) std(X) (Среднее квадратичное отклонение)
Среднее квадратичное отклонение $std(X)$ равно стандартному отклонению $σ$:

$std(X) = σ = 4$


**Ответ**

$M(X) = -2$

$D(X) = 16$

$std(X) = 4$

**а) больше 182 см**

**б) больше 190 см**

**в) от 166 см до 190 см**

**г) от 166 см до 182 см**

**д) от 158 см до 190 см**

**е) не выше 150 см или не ниже 190 см**

**ё) не выше 150 см или не ниже 198 см**

Напишем функцию для определения
-значений для данного $Z$ распределения:

```python
def z_value(height):
    return (height-174)/8
```

```python
from statistics import NormalDist
```

a) Определим $Z$-значениe:

```python
print(z_value(182))
```

метод cdf определяет вероятность того, что случайная величина X будет меньше или равна x, т.е в данном случае $P(X<=182)$

```python
NormalDist().cdf(z_value(182))
```

Однако нам нужна вероятность $P(X>=182)$, поэтому вычтем полученное значение из $1$:

```python
P=1-NormalDist().cdf(z_value(182))
P
```

```python
z=z_value(190)
z
```

Определим $P(X >=190)$

```python
P=1-NormalDist().cdf(z)
P
```

**5)На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?**

Воспользуемся формулой расчета - $Z$ оценки:

$ Z = \frac{X - \mu}{\sigma}  = \frac{190 - 178}{\sqrt{25}} = \frac{12}{5}$

```python
Z=(190-178)/25**(1/2)
Z
```

```python
print(f'Рост человека, равный 190 см, отклоняется от математического ожидания роста в популяции, \n'
      f'в которой M(X) = 178 см и D(X) = 25 кв.см на Z = {Z} сигм.')
```

**Ответ**

$2.4 сигмы$
