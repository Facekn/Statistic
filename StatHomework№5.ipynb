**4) –ï—Å—Ç—å –ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π? –†–æ—Å—Ç –º–∞—Ç–µ—Ä–µ–π 172, 177, 158, 170, 178,175, 164, 160, 169, 165 –†–æ—Å—Ç –≤–∑—Ä–æ—Å–ª—ã—Ö –¥–æ—á–µ—Ä–µ–π: 173, 175, 162, 174, 175, 168, 155, 170, 160**

**1) –ö–æ–≥–¥–∞ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞, –∞ –∫–æ–≥–¥–∞ Z ‚Äì–∫—Ä–∏—Ç–µ—Ä–∏–π?**

Z-–∫—Ä–∏—Ç–µ—Ä–∏–π –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, –∫–æ–≥–¥–∞:

*   –†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–π (–æ–±—ã—á–Ω–æ $ùëõ > 30$).
*   –ò–∑–≤–µ—Å—Ç–Ω–∞ –¥–∏—Å–ø–µ—Ä—Å–∏—è –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.
*   –í—ã–±–æ—Ä–∫–∞ —Å–ª–µ–¥—É–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é.

–ö—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞ (t-–∫—Ä–∏—Ç–µ—Ä–∏–π) –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, –∫–æ–≥–¥–∞:

*   –†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –Ω–µ–±–æ–ª—å—à–æ–π (–æ–±—ã—á–Ω–æ $ùëõ ‚â§ 30$).
*   –î–∏—Å–ø–µ—Ä—Å–∏—è –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞.
*   –í—ã–±–æ—Ä–∫–∞ —Å–ª–µ–¥—É–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –∏–ª–∏ –ø–æ—á—Ç–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É.


**2) –£—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è, —á—Ç–æ —à–∞—Ä–∏–∫–∏ –¥–ª—è –ø–æ–¥—à–∏–ø–Ω–∏–∫–æ–≤, –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã–µ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–º —Å—Ç–∞–Ω–∫–æ–º, –∏–º–µ—é—Ç —Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä 17 –º–º. –ò—Å–ø–æ–ª—å–∑—É—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π —Å Œ±=0,05, –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —ç—Ç—É –≥–∏–ø–æ—Ç–µ–∑—É, –µ—Å–ª–∏ –≤ –≤—ã–±–æ—Ä–∫–µ –∏–∑ n=100 —à–∞—Ä–∏–∫–æ–≤ —Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä –æ–∫–∞–∑–∞–ª—Å—è —Ä–∞–≤–Ω—ã–º 17.5 –º–º, –∞ –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏–∑–≤–µ—Å—Ç–Ω–∞ –∏ —Ä–∞–≤–Ω–∞ 4 –∫–≤. –º–º.**

***–£—Å–ª–æ–≤–∏–µ –∑–∞–¥–∞—á–∏:***

$ùëõ = 100$ (—Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏)

–°—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä —à–∞—Ä–∏–∫–æ–≤ $\bar{x} = 17.5–º–º$

–ó–∞—è–≤–ª–µ–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ${Œº_0} = 17–º–º$

–î–∏—Å–ø–µ—Ä—Å–∏—è $ùúé^2 = 4 –∫–≤.–º–º$ (—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ $ùúé = 2 –º–º$)

–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $Œ± = 0.05$

***–§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –≥–∏–ø–æ—Ç–µ–∑:***

$H_0 : Œº = 17$ (–Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞)

$H_1 : Œº > 17$ (–∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞)

***–í—ã—á–∏—Å–ª–µ–Ω–∏–µ Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏:***

$Z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$

$Z = \frac{17.5 - 17}{2 / \sqrt{100}} = \frac{0.5}{0.2} = 2.5$

***–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è:***

–î–ª—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞ —Å $Œ± = 0.05$, –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $Z_Œ± = 1.645$

***–ü—Ä–∏–Ω—è—Ç–∏–µ —Ä–µ—à–µ–Ω–∏—è –∏ –≤—ã–≤–æ–¥:***

–ü–æ—Å–∫–æ–ª—å–∫—É
$ùëç = 2.5 > 1.645$, –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É.

–£—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä —à–∞—Ä–∏–∫–æ–≤ —Ä–∞–≤–µ–Ω 17 –º–º, –Ω–µ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è.






**3)–ü—Ä–æ–¥–∞–≤–µ—Ü —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å –ø–∞—á–∫–∏ –ø–µ—á–µ–Ω—å—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 200 –≥. –ò–∑ –ø–∞—Ä—Ç–∏–∏ –∏–∑–≤–ª–µ—á–µ–Ω–∞ –≤—ã–±–æ—Ä–∫–∞ –∏–∑ 10 –ø–∞—á–µ–∫. –í–µ—Å –∫–∞–∂–¥–æ–π –ø–∞—á–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∏—Ö –≤–µ—Å–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ. –í–µ—Ä–Ω–æ –ª–∏ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–∞, –µ—Å–ª–∏ —É—á–∏—Ç—ã–≤–∞—Ç—å, —á—Ç–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 99%? (–ü—Ä–æ–≤–µ—Å—Ç–∏ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç.)**

***–£—Å–ª–æ–≤–∏–µ –∑–∞–¥–∞—á–∏:***

$n = 10$ (—Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏)

–í–µ—Å –ø–∞—á–µ–∫: $202, 203, 199, 197, 195, 201, 200, 204, 194, 190$

–ó–∞—è–≤–ª–µ–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: $Œº = 200 –≥$

–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $ùõº = 0.01$ (–¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 99%)

***–§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞ –≥–∏–ø–æ—Ç–µ–∑:***

$H_0 : Œº = 200$ (–Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞)

$H_1 : Œº \neq 200$ (–∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞)

***–í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è:***

$\bar{X} = \frac{202+203+199+197+195+201+200+204+194+190}{10} = 198.5$

$s = \sqrt{\frac{‚àë(X_i - \bar{X})^2}{n-1}}$

–í—ã—á–∏—Å–ª–µ–Ω–∏–µ $ùë†$ –¥–∞–µ—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ $5.16$

***–í—ã—á–∏—Å–ª–µ–Ω–∏–µ t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏:***

$t = \frac{\bar{X} - Œº}{s / \sqrt{n}}$

$t = \frac{198.5 - 200}{5.16 / \sqrt{10}} \approx -0.96$

***–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è:***

–î–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞ —Å $Œ± = 0.01$ –∏ $df = 9$, –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è $t_{Œ±/2,df} = \pm3.25$

***–ü—Ä–∏–Ω—è—Ç–∏–µ —Ä–µ—à–µ–Ω–∏—è:***

–ü–æ—Å–∫–æ–ª—å–∫—É $‚à£t‚à£ = 0.96 < 3.25$, –º—ã –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É

**–í—ã–≤–æ–¥:** –£—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–∞ –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å –ø–∞—á–∫–∏ –ø–µ—á–µ–Ω—å—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 200 –≥, –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è.







**4) –ï—Å—Ç—å –ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π? –†–æ—Å—Ç –º–∞—Ç–µ—Ä–µ–π 172, 177, 158, 170, 178,175, 164, 160, 169, 165 –†–æ—Å—Ç –≤–∑—Ä–æ—Å–ª—ã—Ö –¥–æ—á–µ—Ä–µ–π: 173, 175, 162, 174, 175, 168, 155, 170, 160**

```python
import numpy as np
from scipy import stats

# –†–æ—Å—Ç –º–∞—Ç–µ—Ä–µ–π
mothers_height = np.array([172, 177, 158, 170, 178, 175, 164, 160, 169])

# –†–æ—Å—Ç –¥–æ—á–µ—Ä–µ–π
daughters_height = np.array([173, 175, 162, 174, 175, 168, 155, 170, 160])

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—É—é –¥–ª–∏–Ω—É –º–∞—Å—Å–∏–≤–æ–≤
if len(mothers_height) != len(daughters_height):
    print("–û—à–∏–±–∫–∞: –º–∞—Å—Å–∏–≤—ã —Ä–æ—Å—Ç–∞ –º–∞—Ç–µ—Ä–µ–π –∏ –¥–æ—á–µ—Ä–µ–π –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –¥–ª–∏–Ω—ã.")
else:
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–∞—Ä–Ω–æ–≥–æ t-—Ç–µ—Å—Ç–∞
    t_stat, p_value = stats.ttest_rel(mothers_height, daughters_height)

    # –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    print(f"t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {t_stat}")
    print(f"p-–∑–Ω–∞—á–µ–Ω–∏–µ: {p_value}")

    # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
    alpha = 0.05
    if p_value < alpha:
        print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –µ—Å—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π.")
    else:
        print("–ù–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –Ω–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π.")
```
